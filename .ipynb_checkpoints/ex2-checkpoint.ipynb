```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import copy
```

```python
M = 12
N = 16

def print_h(mat, name):
    print(name + ":\n" + str(mat))
    _ = plt.imshow(mat, interpolation=None)
    plt.colorbar()
    _ = plt.title("$" + name[0] + "_" + name[1] + "$")
    

def create_H(h):
    H = np.zeros((M*N, M*N))
    for i in range(M*N):
        for k in [-1, 0, 1]:
            for l in [-1, 0, 1]:
                j = i-N*k-l
                if 0 <= j < M*N:
                    H[i,j] = h[k+1,l+1]
    return H
```

```python
h1 = (1.0/9.0)*np.ones((3,3))
print_h(h1, "h1")
```

```python
H1 = create_H(h1)
print_h(H1, "H1")
```

```python
h2 = (1.0/8.0)*np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
print_h(h2, "h2")
```

```python
H2 = create_H(h2)
print_h(H2, "H2")
```

```python
h3 = np.transpose(h2)
print_h(h3, "h3")
```

```python
H3 = create_H(h3)
print_h(H3, "H3")
```
