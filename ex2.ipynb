```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import copy
```

```python
M = 12
N = 16

def print_h(mat, name):
#     print(name + ":\n" + str(mat))
    _ = plt.imshow(mat, interpolation=None)
    plt.colorbar()
    _ = plt.title("$" + name[0] + "_" + name[1] + "$")
    

def create_H(h):
    H = np.zeros((M*N, M*N))
    for i in range(M*N):
        for k in [-1, 0, 1]:
            for l in [-1, 0, 1]:
                j = i-N*k-l
                if 0 <= j < M*N:
                    H[i,j] = h[k+1,l+1]
    return H
```

```python
hs = []

hs.append(((1.0/9.0)*np.ones((3,3)), "$h_1$"))
hs.append(((1.0/8.0)*np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]), "$h_2$"))
hs.append((np.transpose(hs[1][0]), "$h_3$"))

fig = plt.figure()
for i in range(len(hs)):
    a = fig.add_subplot(1, 3, i+1)
    implot = plt.imshow(hs[i][0], interpolation=None)
    a.set_title(hs[i][1])
    plt.xticks([]), plt.yticks([])
    plt.colorbar(orientation='horizontal')

plt.show()
```

```python
H1 = create_H(hs[0][0])
H2 = create_H(hs[1][0])
H3 = create_H(hs[2][0])

Hs = []

Hs.append((H1,"$H_1$"))
Hs.append((H2,"$H_2$"))
Hs.append((H3,"$H_3$"))

for i in range(len(hs)):
    fig, a = plt.subplots()
    implot = plt.imshow(Hs[i][0], interpolation=None)
    a.set_title(Hs[i][1])
    a.set_xticks([]), a.set_yticks([])
    plt.colorbar()
    plt.show()
```
